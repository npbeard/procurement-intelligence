```python

import os

from databricks.connect import DatabricksSession
from dotenv import load_dotenv,find_dotenv

load_dotenv(find_dotenv())

DATABRICKS_HOST = os.getenv("DATABRICKS_HOST")
DATABRICKS_TOKEN = os.getenv("DATABRICKS_TOKEN")

spark = DatabricksSession.builder.remote(
    host=DATABRICKS_HOST,
    token=DATABRICKS_TOKEN,
    serverless=True
).getOrCreate()
```

```python
notices = spark.table("capstone.ted.notices")
notices.show(1)
```

```python
extracted_col = notices.select(notices.issue_date).to_date
```

```python
lots = spark.table("capstone.ted.lots")
lots.filter(lots.value > 0).show(5)
```

```python
awarded = spark.table("capstone.ted.award_criteria")
awarded.show(5)
```

```python
org = spark.table("capstone.ted.organizations")
org.show(5)
```
